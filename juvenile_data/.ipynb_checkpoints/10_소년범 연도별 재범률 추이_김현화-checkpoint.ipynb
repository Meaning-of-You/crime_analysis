```python
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fm = mp.font_manager.FontManager()         # 한글화
plt.rcParams['font.family'] = 'Malgun Gothic'
mp.rcParams['axes.unicode_minus'] = False # '-' 에러 방지
```

```python
# 데이터 불러오기
df = pd.read_csv('10_소년범 연도별 재범률 추이.csv') 
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# -를 NaN으로 바꾸기
df.iloc[:,4:] = df.iloc[:,4:].replace('-', np.nan)
```

```python
# 데이터타입 float으로 변경
df.iloc[:,4:] = df.iloc[:,4:].astype(float)
```

```python
# 필요없는 컬럼 제거
df = df.drop(['항목', '단위'], axis=1).copy()
df.head()
```

```python
df_melt = pd.melt(df, id_vars=['범죄별','재범기간및종류별'], var_name='연도', value_name='재범자')
df_melt.head(10)
```

```python
# 인덱스 재설정
df_melt = df_melt.set_index(['연도','범죄별']).copy()
```

```python
df_melt.head()
```

```python
# 피벗테이블
pdf = pd.pivot_table(df_melt,                # 피벗할 데이터프레임
                     index = ['연도', '범죄별'],    # 행 위치에 들어갈 열
                     columns = '재범기간및종류별',    # 열 위치에 들어갈 열
                     values = '재범자')     # 데이터로 사용할 열

pdf
```
