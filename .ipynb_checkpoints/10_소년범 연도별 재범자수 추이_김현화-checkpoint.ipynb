```python
# 데이터 불러오기
df = pd.read_csv('Data/10_소년범 연도별 재범자수 추이_김현화.csv', thousands = ',') 
```

```python
# 컬럼명 변경
df.rename(columns={'자료시점':'연도', '분류':'대분류', '분류.1':'중분류', '분류.2':'소분류'}, inplace=True)
df.head()
```

```python
# 대분류에서 '계'가 들어있는 행만 추출
df_sum = df[df['대분류']=='계'].copy() 
df_sum
```

```python
# 연도, 계, 소년범죄자 계 컬럼만 가져오기
df_sum = df_sum[['연도', '계']].copy()

# 연도를 인덱스로 설정
df_sum.set_index('연도', inplace = True)
```

```python
# 차트 크기 설정
plt.rcParams["figure.figsize"] = (14,8)
# 폰트 크기 설정
plt.rcParams["font.size"] = 13
```

### 1. 소년범 재범자수 추이

```python
df_sum['계'].plot(kind='line', rot=0, figsize=(10, 3))
plt.show()
```

### 2.  범죄유형별 소년 재범자수 추이  
   ###   소분류 기준 상위 5개 범죄: 절도>사기>폭행>상해>도로교통법

```python
# 연도, 소분류, 계 컬럼만 가져오기
df_m = df[['연도', '소분류', '계']]

# 연도를 인덱스로 설정
# df_m.set_index(['연도'], inplace = True)
df_m.head()
```

```python
# 소분류에서 '계'가 들어 있지 않은 행만 추출
df_m = df_m[~df_m.소분류.str.contains('계')].copy()
df_m.head()
```

```python
# 소분류 기준으로 합계 구하기
df_m_sum = df_m.groupby('소분류').sum()

# 내림차순 정렬
df_m_top5 = df_m_sum.sort_values('계',ascending=False).head()
df_m_top5
```

```python
# '연도'를 인덱스로 설정
df_m.set_index("연도", inplace = True)

# 컬럼명 변경
df_m.columns = ['범죄유형', '계']
df_m
```

```python
pdf_m = pd.pivot_table(df_m,                # 피벗할 데이터프레임
                     index = '범죄유형',    # 행 위치에 들어갈 열
                     columns = '연도',    # 열 위치에 들어갈 열
                     values = '계')     # 데이터로 사용할 열
pdf_m
```

```python
# 연도별 합계 컬럼 추가
pdf_m['연도별_합계'] = pdf_m.sum(axis=1)
pdf_m.head()
```

```python
# 2014년부터 2019년까지 합계 기준 내림차순 정렬 => 상위 5개 범죄 유형.: 절도, 사기, 폭행, 상해, 도로교통법 (위반)
result = pdf_m.sort_values('연도별_합계',ascending=False).head()
result
```

```python
# 연도를 x축으로 하는 그래프 보기 위해 행렬 전치 
result_T = result.head().T
result_T
```

```python
# 연도별_합계 행 제거하고 연도별 상위 5개 소년범죄 추이 선 그래프로 시각화
remove_total = result_T.drop(['연도별_합계'])
remove_total.plot(figsize=(15, 5))
plt.ylabel('소년 재범자수')
```
